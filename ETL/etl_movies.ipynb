```python
import numpy as np
import pandas as pd
```

```python
#abro el csv con pandas
movies= pd.read_csv('../Datos/movies_dataset.csv')
movies.info()
```

```python
#selecciona la columna y la convierto en un diccionario con el fin de prepararla para el siguiente paso
nombres= movies['belongs_to_collection']
nombres=dict(nombres)
```

```python
# Busco iterar toda la serie, verificar que hay diccionario para tomar el valor de la clave 'name', contemplando los njulos
import json
import pandas as pd

def obtener_nombres(series):
    for nombre_dict_str in series:
        if pd.isnull(nombre_dict_str):
            yield None
        else:
            nombre_json = str(nombre_dict_str).replace("'", '"')
            try:
                nombre_dict = json.loads(nombre_json)
                if isinstance(nombre_dict, dict):  # Verificar que sea un diccionario
                    nombre = nombre_dict.get('name')
                    if nombre:
                        yield nombre
            except json.JSONDecodeError:
                pass

nombres = obtener_nombres(movies['belongs_to_collection'])

#guardo los datos en la lista collection
collection = []
for nombre in nombres:
   collection.append(nombre)

collection
```

```python
# Agregar el DataFrame collection al DataFrame existente
df_collection= pd.DataFrame(collection, columns=['collection'])
movies = pd.concat([movies, df_collection], axis=1)
movies.head(5)
```

```python
#en cada fila de genres aplico la funcion
movie_genres = movies['genres'].apply(lambda x: [d['name'] for d in eval(x)] if pd.notnull(x) else [])
movies['genres']= movie_genres
```

```python
#expresion regular para extraer los name referidos a los paises
movies['production_countries']= movies['production_countries'].str.extract(r"'name': '([^']*)'")
```

```python
movies['production_companies']=movies['production_companies'].str.extract(r"'name': '([^']*)'")
```

```python
movies["spoken_languages"]=movies["spoken_languages"].str.extract(r"'name': '([^']*)'")
```

```python
#b)Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0
movies['revenue'].fillna(0,inplace=True)
movies['budget'].fillna(0,inplace=True)
```

```python
#c) Eliminar valores nulos en release_date
movies= movies.dropna(subset=['release_date'])
movies['release_date'].isnull().sum()
```

```python
#d) De haber fechas, deberá tener el formato AAAA-mm-dd, además deberá crear la columna release_yeardonde extraerán el año de la fecha de estreno.
#hay algunos formatos bien, pero pasamos to_datetime para igual todos los datos
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
movies['release_date'] = pd.to_datetime(movies['release_date']).dt.strftime('%Y-%m-%d')#formato solicitado
```

```python
#extraigo el año de release_date y lo paso como un DF
año = movies['release_date'].str.slice(0, 4)
año= list(año)
df_año= pd.DataFrame(año, columns=['release_year'])  # 45376 filas
```

```python
#hago lo mismo con release_date. Para luego sumarlo como una nueva columna modificada
date= movies['release_date']
date= list(date)
df_date= pd.DataFrame(date, columns=['release_date'])
df_date= df_date.dropna() #45376 filas
```

```python
#Al comprobar que tiene la misma cantidad de filas y no hay nulos. Elimino el release_date.
movies.drop(columns=['release_date'], inplace=True)
movies.columns
```

```python
#Ahora concateno al DF original, las nuevas columnas
movies = pd.concat([movies, df_date, df_año], axis=1)
```

```python
#e)Cree la columna con el retorno de inversión, llame return con los campos revenuey budget, dividiendo estas dos últimas revenue / budget, cuando no haya datos disponibles para calcularlo, deberá tomar el valor 0.
#me tira error! debo convertir los datos a float
retorno = ((pd.to_numeric(movies['revenue'], errors='coerce')) /  (pd.to_numeric(movies['budget'], errors='coerce')))
retorno = pd.DataFrame(retorno, columns=['return'])
retorno.fillna(0, inplace=True)
retorno = retorno.replace(np.inf,0)
retorno
```

```python
movies = pd.concat([movies, retorno], axis=1)
movies.head(5)
```

```python

# Contar los valores NaN en cada fila
num_nan = movies.isnull().sum(axis=1)

# Filtro y elimino las filas que tienen más de 15 NaN (15, porque conclui que con menos puede igual ser relevante )
movies = movies.dropna(thresh=movies.shape[1]-15)

movies
```
